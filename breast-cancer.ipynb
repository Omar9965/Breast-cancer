## Splitting Data

```python
X = df.drop("diagnosis", axis=1)
y = df["diagnosis"] 
plt.figure(figsize=(30, 15))
X.boxplot()
```

```python
z = np.abs(stats.zscore(X))


threshold_z = 3

outlier_indices = np.where(z > threshold_z)[0]
X_no_outliers = X.drop(outlier_indices)
print("Original DataFrame Shape:",X.shape)
print("DataFrame Shape after Removing Outliers:", X_no_outliers.shape)
```

```python
scaler = MinMaxScaler()
X_clean = scaler.fit_transform(X) 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_clean, y, test_size=0.15, shuffle=True ,random_state=42)
```

## Random Forest Hyper parameter tuning

```python
random_forest = RandomForestClassifier()
params = {"n_estimators" : np.arange(100, 300, 50), "max_depth" : np.arange(2, 10, 2)} 
grid_model = GridSearchCV(estimator= random_forest, param_grid=params, scoring="neg_mean_squared_error", cv=5, n_jobs=-1)
grid_model.fit(X_train, y_train)
print(f"Grid model best paraemeters {grid_model.best_params_}")
best_model = grid_model.best_estimator_
best_model.fit(X_train, y_train)
```

```python
last_pred = best_model.predict(X_test)
print(f"Model accuracy: {metrics.accuracy_score(y_test, last_pred)}")
```

```python
conf_matrix = metrics.confusion_matrix(y_test, last_pred)
sns.heatmap(conf_matrix, annot=True, cmap=sns.color_palette("mako", as_cmap=True))
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```
